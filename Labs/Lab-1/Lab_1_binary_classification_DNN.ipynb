```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# Load and preprocess the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalize the image data
train_images = train_images / 255.0
test_images = test_images / 255.0

# Convert labels to binary (1 for digit '5', 0 for all others)
train_labels = (train_labels == 5).astype(float)
test_labels = (test_labels == 5).astype(float)

# Define the neural network model
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),  # Flatten the 28x28 images into 784 input features
    layers.Dense(10, activation='relu'),   # First hidden layer with 10 neurons
    layers.Dense(8, activation='relu'),    # Second hidden layer with 8 neurons
    layers.Dense(8, activation='relu'),    # Third hidden layer with 8 neurons
    layers.Dense(4, activation='relu'),    # Fourth hidden layer with 4 neurons
    layers.Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Model summary to see the structure
model.summary()

# Train the model
history = model.fit(train_images, train_labels, epochs=50, validation_split=0.2)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(test_images, test_labels)

print(f"Test Accuracy: {test_acc*100:.2f}%")

# Save the model
model.save('my_mnist_model.keras')

# Load the model (for future use)
loaded_model = tf.keras.models.load_model('my_mnist_model.keras')

```
